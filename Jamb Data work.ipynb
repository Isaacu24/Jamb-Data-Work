## T-test to determine if there is a significant difference between scores for male and female

-Null hypothesis: There is no significant difference in the average Jamb scores of male and female students

-Alternative hypothesis: There is significant difference in the average Jamb scores of male and female students

```python
male_students=data[data['Gender']=='Male']['JAMB_Score']
female_students=data[data['Gender']=='Female']['JAMB_Score']
```

```python
t_stat, p_value=stats.ttest_ind(male_students, female_students)
t_stat, p_value
```

### Interpretation
- A t-test value of -0.86 indicates that the difference between the means of the two groups (gender, and JAMB scores) is relatively small compared to the variability within those groups. The negative value suggests that the mean of the first group is slightly lower than that of the second group, but the difference is minimal.

- The p-value of 0.38 is much higher than the common significance level of 0.05. This means we fail to reject the null hypothesis (H0), which states that there is no significant difference in the average Jamb scores of male and female students

- Practical Interpretation: The observed difference (indicated by the t-test of -0.86) is likely due to random variability rather than a true effect of gender. In practical terms, this suggests that gender does not have a meaningful impact on JAMB scores in this data set.


## Attendance rate between Public School and Private School

-Null hypothesis: There is no significant difference in the average Attendance rate of public and private school students

-Alternative hypothesis: There is significant difference in the average Attendance rate of public and private school students

```python
public_attendance=data[data['School_Type']=='Public']['Attendance_Rate']
private_attendance=data[data['School_Type']=='Private']['Attendance_Rate']
```

```python
t_stat, p_value=stats.ttest_ind(public_attendance, private_attendance)
t_stat, p_value
```

### Interpretation
- The p-value of 0.000077 (or 7 x 10^-5) is extremely low, much lower than the common significance level of 0.05. This indicates that the result is statistically significant
- Practical Interpretation: Despite the low t-test result, the very small p-value indicates strong evidence against the null hypothesis(rejecting the null hypothesis). It suggests that the difference in attendance rate between the school types is unlikely to be due to random chance.

## Parent Involvement and Average Jamb Score (CHI2- TEST)

- Null hypothesis: There is no association between parent involvement and JAMB scores. This means that the distribution of test scores is the same regardless of the level of parent involvement.
- Alternative hypothesis: There is an association between parent involvement and JAMB scores. This suggests that the distribution of test scores differs based on the level of parent involvement.

```python
data['bolean_jambscore']=data['JAMB_Score']>data['JAMB_Score'].mean()
data
```

```python
par_sco_table= pd.crosstab(data['Parent_Involvement'], data['bolean_jambscore'])
par_sco_table
```

```python
chi2_stat, p, dof, expected = stats.chi2_contingency(par_sco_table)
chi2_stat, p, dof, expected
```

### Interpretation
- A Chi-square statistic of 72.43 indicates a significant difference between the observed frequencies and the expected frequencies under the null hypothesis. This high value suggests that there is a strong association between parent involvement and JAMB scores
- The p-value of 1.86 x 10^-16 is extremely small and far below the common significance level of 0.05. This means we can reject the null hypothesis, which typically states that there is no association between the two categorical variables (parent involvement and JAMB scores)
- The degrees of freedom in a Chi-square test are calculated based on the number of categories in our data. In this case, having 2 degrees of freedom is typical when comparing three groups or categories in your contingency table
- Practical Implication: The results suggest that different levels or types of parent involvement are associated with variations in JAMB scores. This could imply that increasing parent involvement may lead to better test performance, but further investigation would be needed to explore the nature of this relationship.
